# ✈️ UAV Waypoint Demo Notebook
Plan and simulate a UAV mission using a simple list of GPS coordinates.

```python
# Sample waypoints for visualization
import matplotlib.pyplot as plt

waypoints = [
    (37.4289, -122.1760),
    (37.4295, -122.1750),
    (37.4290, -122.1740)
]

lats, lons = zip(*waypoints)
plt.plot(lons, lats, marker='o')
plt.title("Waypoint Path")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```
